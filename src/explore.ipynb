# APIs

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

## Obtenemos los datos de la API

```python
load_dotenv()
```

## Obtenemos credenciales de Spotify

```python
client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")
```

## Inicializamos la biblioteca Spotipy

```python
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spotify = spotipy.Spotify(auth_manager=auth_manager)
```

## Realizamos solicitudes a la API

```python
id = 'spotify:artist:53XhwfbYqKCa1cC15pYq2q'
top_songs = spotify.artist_top_tracks(id)['tracks']
top_songs
```

## Transformamos los datos a Pandas DataFrame

```python
df = pd.DataFrame(top_songs)
df
```

## Ordenamos por popularidad creciente y damos el top 3

```python
sorted_df = df.sort_values('popularity')
sorted_df.head(3)
```

## código alternativo para crear el dataframe de manera más limpia y bonita


```python
artist_id = id
top_songs = spotify.artist_top_tracks(artist_id)['tracks']

top_songs_df = pd.DataFrame(top_songs)
top_songs_df['duration_min'] = top_songs_df['duration_ms'] / 1000 / 60
top_songs_df_filter = top_songs_df[['name', 'popularity', 'duration_min']]

top_songs_df_filter
```

## Analizamos la relación estadistica entre duración y popularidad

```python
plt.scatter(df['duration_ms'], df['popularity'])
plt.title('Relación entre la popularidad y la duración')
plt.xlabel('Duración en ms')
plt.ylabel('Popularidad de la canción')
plt.show
```

## Conclusiones
>No veo ninguna correlación clara entre la popularidad de la canción y su duración ya que parece que los datos estan muy dispersos entre el top 10 de las más populares de Imagine Dragons, lo que si que podemos afirmar es que tenemos claramente canciones más cortas y a su vez más populares que otras ya que la más larga se encuentra por debajo de varias más populares.
