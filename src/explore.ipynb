# APIs

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

## Obtenemos los datos de la API

```python
load_dotenv()
```

## Obtenemos credenciales de Spotify

```python
client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")
```

## Inicializamos la biblioteca Spotipy

```python
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spotify = spotipy.Spotify(auth_manager=auth_manager)
```

## Realizamos solicitudes a la API

```python
id = 'spotify:artist:53XhwfbYqKCa1cC15pYq2q'
top_songs = spotify.artist_top_tracks(id)['tracks']
top_songs
```

## Transformamos los datos a Pandas DataFrame

```python
df = pd.DataFrame(top_songs)
df
```

## Ordenamos por popularidad creciente y damos el top 3

```python
sorted_df =df.sort_values('popularity')
sorted_df.head(3)
```

## Analizamos la relacin estadistica entre duracin y popularidad

```python
plt.scatter(df['duration_ms'], df['popularity'])
plt.title('Relacin entre la popularidad y la duracin')
plt.xlabel('Duracin en ms')
plt.ylabel('Popularidad de la cancin')
plt.show
```

## Conclusiones
>No veo ninguna correlacin clara entre la popularidad de la cancin y su duracin ya que parece que los datos estan muy dispersos entre el top 10 de las ms populares de Imagine Dragons, lo que si que podemos afirmar es que tenemos claramente canciones ms cortas y a su vez ms populares que otras ya que la ms larga se encuentra por debajo de varias ms populares. Por lo que se ve en la grafica tamin, podra ser interesante eliminar de la grafica la cancin ms larga ya que est muy alejada del resto. Lo hago y doy otras conclusiones finales.

```python
plt.scatter(df['duration_ms'], df['popularity'])
plt.title('Relacin entre la popularidad y la duracin normalizando los datos')
plt.xlabel('Duracin en ms')
plt.ylabel('Popularidad de la cancin')
plt.xlim(0,240000)
plt.show
```
